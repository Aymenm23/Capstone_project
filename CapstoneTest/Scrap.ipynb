```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Grouping by features and summing the 'yield' column
df_grouped = thai_filtered.groupby('Item').sum().reset_index()

# Splitting data into features (X) and target variable (y)
X = df_grouped.drop(columns=['Yield', 'Item', 'Area'])  # Features
y = df_grouped['Yield']  # Target variable (sum of 'yield' for each crop)

# Splitting the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Instantiate linear regression model
model = LinearRegression()

# Fit the model on the training data
model.fit(X_train, y_train)

# Predict on the testing data
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
print("Mean Squared Error:", mse)
```

```python
avg = avg[(avg['avg_yield'] != 0) & (avg['avg_area_harvested'] != 0)]
```

```python
avg
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

# Step 1: Prepare the data
avg_yield_per_crop = avg.groupby(['Country', 'Crop'])['avg_yield'].mean().reset_index()

# One-hot encode the 'Country' column
encoder = OneHotEncoder(sparse=False)
encoded_countries = encoder.fit_transform(avg_yield_per_crop[['Country']])
encoded_df = pd.DataFrame(encoded_countries, columns=encoder.get_feature_names_out(['Country']))

# Concatenate the encoded country features with the DataFrame
avg_yield_per_crop_encoded = pd.concat([avg_yield_per_crop, encoded_df], axis=1)

# Step 2: Train the model for each crop
models = {}
for crop in avg_yield_per_crop['Crop'].unique():
    # Filter data for the current crop
    crop_data = avg_yield_per_crop_encoded[avg_yield_per_crop_encoded['Crop'] == crop]
    
    # Prepare features (country) and target (yield)
    X = crop_data.drop(['Country', 'Crop', 'avg_yield'], axis=1)
    y = crop_data['avg_yield']
    
    # Train the linear regression model
    model = LinearRegression()
    model.fit(X, y)
    
    # Store the trained model
    models[crop] = model

# Step 3: Predict total average yield for each country
predicted_yields = {}
for country in avg_yield_per_crop['Country'].unique():
    total_yield = 0
    for crop, model in models.items():
        # Prepare features for prediction (one-hot encoded country)
        country_encoded = encoder.transform([[country]])
        
        # Predict yield for the current country using the corresponding model
        yield_prediction = model.predict(country_encoded)
        total_yield += yield_prediction[0]
    # Store the predicted total yield for the current country
    predicted_yields[country] = total_yield

# Print the predicted total average yield for each country
for country, total_yield in predicted_yields.items():
    print(f"Predicted total average yield for {country}: {total_yield}")
```

```python
# Filter data for Yemen and Thailand
yemen_data = avg[avg['Country'] == 'Yemen']
thailand_data = avg[avg['Country'] == 'Thailand']

# Calculate total average yield for Yemen
yemen_total_yield = yemen_data['avg_yield'].sum()

# Calculate total average yield for Thailand
thailand_total_yield = thailand_data['avg_yield'].sum()

print("Actual total average yield for Thailand:", thailand_total_yield)
print("Actual total average yield for Yemen:", yemen_total_yield)
```

```python
# Define actual total average yields for Yemen and Thailand
actual_yields = {
    'Yemen': round(yemen_total_yield,2),
    'Thailand': thailand_total_yield
}

# Define predicted total average yields obtained from the linear regression model
predicted_yields = {
    'Yemen': round(predicted_yields['Yemen'],2),
    'Thailand': predicted_yields['Thailand']
}

# Print actual vs predicted yields
print("Actual vs Predicted Total Average Yields:")
for country in actual_yields.keys():
    print(f"{country}:")
    print(f"  Actual: {actual_yields[country]}")
    print(f"  Predicted: {predicted_yields[country]}")
    print(f"  Difference: {actual_yields[country] - predicted_yields[country]}")
    print()
```
